```python
import pandas as pd
import matplotlib.pyplot as plt
import io
from IPython.display import display
import ipywidgets as widgets

def upload_score_data():
    # 사용자가 파일을 업로드하도록 합니다.
    from google.colab import files
    uploaded = files.upload()

    # 업로드된 파일을 읽어옵니다.
    file_name = next(iter(uploaded))
    data = pd.read_csv(io.BytesIO(uploaded[file_name]), encoding='cp949')
    return data

def analyze_score_data(data):
    # 전체 학생 수
    total_students = data.shape[0]

    # 평균 점수 계산
    average_score = data['점수'].mean()

    # 점수를 내림차순으로 정렬
    sorted_scores = data['점수'].sort_values(ascending=False).reset_index(drop=True)

    # 등급별 인원 수 계산
    level_thresholds = {
        1: round(total_students * 0.04),
        2: round(total_students * 0.07),
        3: round(total_students * 0.12),
        4: round(total_students * 0.17),
        5: round(total_students * 0.20),
        6: round(total_students * 0.17),
        7: round(total_students * 0.12),
        8: round(total_students * 0.07),
        9: round(total_students * 0.04)
    }

    # 등급별 점수 기준 계산
    thresholds = {}
    start_idx = 0
    for level, count in level_thresholds.items():
        end_idx = start_idx + count - 1
        thresholds[level] = sorted_scores.iloc[end_idx]
        start_idx = end_idx + 1

    return thresholds, average_score

def determine_grade(thresholds, score):
    # 입력된 점수가 어떤 등급에 속하는지 확인
    for level in sorted(thresholds.keys()):
        if score >= thresholds[level]:
            return level
    return 9

def determine_difficulty(average_score, standard_deviation):
    # 시험 난이도를 판별합니다.
    if average_score > 70 and standard_deviation < 15:
        return "Easy"
    elif average_score < 50 and standard_deviation > 15:
        return "Hard"
    else:
        return "Medium"

def visualize_grade_distribution(thresholds, user_score, average_score, standard_deviation, difficulty):
    # 등급별 점수 기준을 그래프로 시각화
    levels = list(thresholds.keys())
    scores = [thresholds[level] for level in levels]

    plt.figure(figsize=(10, 6))
    plt.plot(levels, scores, marker='o', linestyle='-', label='Grade Boundaries')
    plt.xlabel('Grade Level')
    plt.ylabel('Score')
    plt.title(f'Grade Boundaries by Score - Difficulty: {difficulty}')

    # 평균 점수를 수평선과 텍스트로 표시
    plt.axhline(y=average_score, color='blue', linestyle='--', label='Average Score')
    plt.text(x=1.1, y=average_score, s=f'Avg: {average_score:.2f}', color='blue', va='center')

    # 사용자 점수에 해당하는 등급 표시
    user_grade = determine_grade(thresholds, user_score)
    plt.axhline(y=user_score, color='red', linestyle='--', label=f'Your Score: {user_score}')
    plt.text(x=1.1, y=user_score, s=f'Your Score: {user_score}', color='red', va='center')

    plt.xticks(levels)
    plt.gca().invert_xaxis()  # 등급이 높은 순서대로 표시되도록 X축 반전
    plt.legend()
    plt.grid(True)
    plt.show()

# 사용 예시:
# 1. 점수 데이터를 업로드합니다.
data = upload_score_data()

# 2. 점수 데이터를 분석합니다.
thresholds, average_score = analyze_score_data(data)
standard_deviation = data['점수'].std()
difficulty = determine_difficulty(average_score, standard_deviation)

# 3. 특정 점수를 입력하여 해당 점수가 어떤 등급인지 확인합니다.
score_input = widgets.FloatText(
    value=85.0,
    description='Your Score:',
    disabled=False
)

display(score_input)

# 4. 입력된 점수에 따라 등급을 계산하고 결과를 출력합니다.
def on_score_change(change):
    user_score = change['new']
    user_grade = determine_grade(thresholds, user_score)
    print(f"Your score of {user_score} falls into Level {user_grade}.")
    print(f"Average Score: {average_score:.2f}")
    print(f"Standard Deviation: {standard_deviation:.2f}")
    print(f"Difficulty Level: {difficulty}")
    visualize_grade_distribution(thresholds, user_score, average_score, standard_deviation, difficulty)

score_input.observe(on_score_change, names='value')
```
